```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/heart.csv")
df.head()
```

```python
df.info()
```

```python
#check for NAN values
df.isnull().sum()
```

```python
#general analysis
df.shape
df.describe()
```

```python
#target distribution
df.target.value_counts()
```

```python
#show correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
#plot target dist.
#sns.countplot(df.target,hue='sex',palette=['green','red'])
#sns.countplot(x=df.target,data=df,palette=['#1CA53B','red'])
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
#seaborn.pairplot(df,hue='size)
plt.xticks([0,1],labels=['Not Disease','Disease'])
plt.title("Target Distribution");
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','orange'],edgecolor="k")
plt.legend(["Haven't Disease","Have Disease"])
plt.title("Heart Disease Frequency for Ages")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#sns.set_style("whitegrid")
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='blue',label='Haven\'t Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='red',label='Have Disease')
plt.legend()
plt.title("Heart Disease Frequency for Ages")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
